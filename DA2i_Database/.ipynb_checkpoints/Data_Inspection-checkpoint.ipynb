```python
ind_internet = DA2i.loc[(DA2i.Country == 'Senegal') & (DA2i.Name == 'ind.internet')]
ind_internet = ind_internet.sort_values('Year', ascending=False)

ind_internet = pd.DataFrame(data=ind_internet, columns=['Year','value'])
ind_internet.set_index('Year', inplace=True)
ind_internet = ind_internet.rename(columns={'value':'% of Individuals Using Internet'})

a = []
a.append(ind_internet['% of Individuals Using Internet'].iloc[0] - ind_internet['% of Individuals Using Internet'].iloc[1])
a.append(ind_internet['% of Individuals Using Internet'].iloc[1] - ind_internet['% of Individuals Using Internet'].iloc[2])
a.append(ind_internet['% of Individuals Using Internet'].iloc[2] - ind_internet['% of Individuals Using Internet'].iloc[3])
a.append(ind_internet['% of Individuals Using Internet'].iloc[3] - ind_internet['% of Individuals Using Internet'].iloc[4])

ind_internet = ind_internet[0:4]
ind_internet['% Change from Prev. Year'] = a
ind_internet.astype('int')
```
